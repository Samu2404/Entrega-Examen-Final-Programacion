```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

```

### 1.Carga de los datos y binarizacion del target

```python
# Cargar los datos desde el archivo CSV
df = pd.read_csv("epidata.csv")


# Reclasificar   la columna 1 como individuo individuo
df.rename(columns={'Unnamed': 'individuo'}, inplace=True)


# Renombramos la columna "y" como label para claridad
df.rename(columns={'y': 'label'}, inplace=True)


# Creamos una columna binaria: 1 = epileptic seizure, 0 = no seizure (clases 2,3,4 y 5)
df['epilepsy_binary'] = df['label'].apply(lambda x: 1 if x == 1 else 0)

# Mostrar primeras filas
print("\nPrimeras filas del DataFrame organizado:")
print(df.head().to_string())
```

### 2. Separacion de datos

```python
# Datos de entrada 
x = df.drop(columns=['individuo', 'label', 'epilepsy_binary'])

# Datos de salida
y = df['epilepsy_binary']
```

### 3. Escalado de datos

```python
# Escalar los datos de entrada a un rango entre 0 y 1
escalador_minmax=MinMaxScaler()
datos_escalado = escalador_minmax.fit_transform(x)
datos_escalado = pd.DataFrame(datos_escalado, columns=x.columns)
(datos_escalado.head())
```

### 4. Exportacion de los datos procesados

```python
# Exportar los datos de entrada normalizados
datos_escalado.to_csv('datos_entrada_normalizados.csv', index=False)

# Exportar los datos de salida (etiquetas)
y.to_csv('datos_salida.csv', index=False, header=True)
```
